```python
# Define a list of old and new variable names
variable_mappings = [
    ('old_name_1', 'new_name_1'),
    ('old_name_2', 'new_name_2'),
    ('old_name_3', 'new_name_3'),
]

# Create a dictionary with the original data
data = {
    'old_name_1': 10,
    'old_name_2': 20,
    'old_name_3': 30,
}

# Create a new dictionary with the updated variable names
new_data = {}

# Loop through the variable mappings and update the new dictionary
for old_name, new_name in variable_mappings:
    new_data[new_name] = data[old_name]

# Print the new dictionary
print(new_data)
```
