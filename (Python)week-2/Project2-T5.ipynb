```python
import json

def add_transaction(user_data):
    amount = float(input("Enter the transaction amount: "))
    category = input("Enter the transaction category: ")
    date = input("Enter the transaction date (YYYY-MM-DD): ")

    transaction = {
        "amount": amount,
        "category": category,
        "date": date
    }

    user_data["transactions"].append(transaction)
    save_data(user_data)
    print("Transaction added successfully!")

def view_transactions(user_data):
    sort_option = input("Sort transactions by (date/category): ")

    if sort_option.lower() == "date":
        sorted_transactions = sorted(user_data["transactions"], key=lambda t: t["date"])
    elif sort_option.lower() == "category":
        sorted_transactions = sorted(user_data["transactions"], key=lambda t: t["category"])
    else:
        print("Invalid sort option. Transactions will be displayed in the order they were entered.")
        sorted_transactions = user_data["transactions"]

    print("Transactions:")
    for transaction in sorted_transactions:
        print(f"- Amount: {transaction['amount']}, Category: {transaction['category']}, Date: {transaction['date']}")

def generate_reports(user_data):
    total_expenses = sum(
        transaction["amount"] for transaction in user_data["transactions"] if transaction["amount"] < 0
    )
    total_income = sum(
        transaction["amount"] for transaction in user_data["transactions"] if transaction["amount"] > 0
    )

    print(f"Total Expenses: {total_expenses}")
    print(f"Total Income: {total_income}")

    print("Spending by Category:")
    categories = set(transaction["category"] for transaction in user_data["transactions"])
    for category in categories:
        category_expenses = sum(
            transaction["amount"]
            for transaction in user_data["transactions"]
            if transaction["category"] == category
        )
        print(f"- {category}: {category_expenses}")

    print("Transactions in Each Category:")
    for category in categories:
        category_transactions = [
            transaction
            for transaction in user_data["transactions"]
            if transaction["category"] == category
        ]
        print(f"- {category}:")
        for transaction in category_transactions:
            print(f"  - Amount: {transaction['amount']}, Date: {transaction['date']}")

    print("Average Transaction Value by Category:")
    for category in categories:
        category_transactions = [
            transaction
            for transaction in user_data["transactions"]
            if transaction["category"] == category
        ]
        transaction_values = [transaction["amount"] for transaction in category_transactions]
        average_value = sum(transaction_values) / len(transaction_values)
        print(f"- {category}: {average_value}")

def save_data(user_data):
    with open(user_data["filename"], "w") as file:
        json.dump(user_data, file)

def load_data(filename):
    try:
        with open(filename, "r") as file:
            return json.load(file)
    except FileNotFoundError:
        return {
            "filename": filename,
            "transactions": []
        }

def main():
    print("-- Personal Expenses and Income Tracker --")
    while True:
        print("\n1. Log in")
        print("2. Create a new account")
        print("3. Exit")

        choice = input("Enter your choice (1-3): ")

        if choice == "1":
            username = input("Enter your username: ")
            filename = f"{username}.json"
            user_data = load_data(filename)
            print(f"Welcome, {username}!")
            while True:
                print("\n1. Add Transaction")
                print("2. View Transactions")
                print("3. Generate Reports")
                print("4. Log out")

                choice = input("Enter your choice (1-4): ")

                if choice == "1":
                    add_transaction(user_data)
                elif choice == "2":
                    view_transactions(user_data)
                elif choice == "3":
                    generate_reports(user_data)
                elif choice == "4":
                    save_data(user_data)
                    print("Logging out...")
                    break
                else:
                    print("Invalid choice. Please try again.")

        elif choice == "2":
            username = input("Create a new username: ")
            filename = f"{username}.json"
            user_data = load_data(filename)
            print(f"Account created successfully for {username}!")
            while True:
                print("\n1. Add Transaction")
                print("2. View Transactions")
                print("3. Generate Reports")
                print("4. Log out")

                choice = input("Enter your choice (1-4): ")

                if choice == "1":
                    add_transaction(user_data)
                elif choice == "2":
                    view_transactions(user_data)
                elif choice == "3":
                    generate_reports(user_data)
                elif choice == "4":
                    save_data(user_data)
                    print("Logging out...")
                    break
                else:
                    print("Invalid choice. Please try again.")

        elif choice == "3":
            print("Exiting...")
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```
